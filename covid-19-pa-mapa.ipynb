```python
df = pd.read_csv("../input/covid-19-casos-pa/Covid-19_PA - Pgina2.csv")
```

```python
mapa = folium.Map(location = [-9.412239,-38.231131], tiles = 'openstreetmap', zoom_start = 13)
```

```python
for i in range(0, len(df)):
    folium.Circle(
    location = [df.iloc[i]['Lat'], df.iloc[i]['Long']], color = '#F80000', fill = '#F80000',
    tooltip = str(df.iloc[i]['Bairro']) + ': ' + str(df.iloc[i]['Casos']) + ' casos', radius = int(df.iloc[i]['Casos']*20)).add_to(mapa)    
    
```
